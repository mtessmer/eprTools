```python
from DEER import DEER_Spec
import matplotlib.pyplot as plt
```

```python
myspc = DEER_Spec.from_file('180529_ExoU_D133R1_S142A_G439R1_150uM_3500ns.DTA')
```

```python
myspc.phase()
myspc.zero_time()
myspc.correct_background()
myspc.get_fit()
```

```python
plt.plot(myspc.time, myspc.dipolar_evolution)
plt.plot(myspc.fit_time, myspc.fit)
plt.show()
plt.plot(myspc.r, myspc.P)
plt.show()
```
